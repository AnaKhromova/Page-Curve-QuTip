## Introduction

### Unitary and Non-Unitary Isometric Time Evolution

**Unitary** time evolution is a fundamental concept in quantum mechanics. It means that the state of a quantum system evolves over time according to a **unitary operator**. This operator is reversible and preserves the total probability (i.e., the sum of the probabilities of all possible outcomes remains 1). It is a way to describe how a **closed** quantum system changes over time without losing information.

**Non-unitary isometric time evolution**, on the other hand, is a bit more complex. It's used when dealing with **open quantum systems**, which can interact with their environment. In this case, the evolution of the system is described by an isometric but non-unitary operator. This means that while the transformation still preserves the norm (i.e., the total probability remains 1), it is not necessarily reversible. This **non-reversibility** reflects the fact that information can be lost to the environment in an **open system**.

A **quantum open system** refers to a quantum system that interacts with an external environment (often called a "bath" or "reservoir"). Because of this interaction, the system is not isolated, and its dynamics are influenced by the environment. This leads to phenomena such as decoherence (loss of quantum coherence due to interaction with the environment), dissipation (loss of energy from the system to the environment) and noise (random fluctuations introduced by the environment). In open quantum systems, the evolution of the system is typically **non-unitary** because information (e.g., quantum states or energy) is lost to the environment. This is described mathematically using tools like the e.g. master equation like **Lindblad equation**. **Non-unitary isometric evolution** refers to a specific type of mathematical description used in quantum mechanics. The combined system (system + environment) evolves **unitarily**, but the evolution of the system alone is non-unitary because it is "traced over" the environment (i.e., the environment is treated as an external degrees of freedom that is not observed).

In **open quantum systems**, the evolution is typically **non-unitary** because the system can exchange information and energy with its environment. This is in contrast to **closed systems**, where the evolution is **unitary** and no information is lost. When we consider the **combined system** of the quantum system and its environment, the evolution is unitary. However, when we focus only on the system and "trace over" the environment (i.e., average over all possible states of the environment), the resulting evolution for the system alone is **non-unitary**. This reflects the fact that we're not keeping track of the environment's degrees of freedom, which can carry away information from the system.

The **Page curve** is a concept from quantum information theory that describes how the **entanglement entropy** of a subsystem evolves over time when a **pure state** is subjected to **unitary evolution**. In simpler terms, it shows how the amount of information shared between two parts of a system changes as the system evolves. The curve is named after Don Page, who first described it in the context of **black hole evaporation**. The **Page curve** typically has a characteristic shape: it starts at zero, rises to a maximum value, and then decreases back to zero. This behavior is thought to be a consequence of the subsystem becoming increasingly **entangled** with the rest of the system, and then becoming less entangled as the system approaches a pure state again.

The **Page curve**, which describes the entropy dynamics of a **black hole**, provides a framework for understanding how information is preserved during **black hole evaporation**. According to the Page curve, the entropy of a black hole initially increases as it emits Hawking radiation, reaching a maximum value at the Page time. After the **Page time**, the entropy starts to decrease: the emitted radiation carries the information away. This decrease in entropy suggests that the evaporation process is **unitary** and that information is not lost, but rather encoded in the **entanglement** between the black hole and the radiation.

### The Lindblad Master Equation
The Lindblad equation (master equation in Lindblad form, quantum Liouvillian, or Lindbladian) is a popular Markovian quantum master equation that is widely used in quantum optics, quantum information, and open quantum systems. Nevertheless, it is still an approximation to the full quantum dynamics and may not capture all the relevant physics.

```python
# Define parameters
Omega =1  # Rabi frequency represents the strength of the interaction between the two states.
hbar=1
sx = sigmax() # Pauli X matrix
Delta=0.1 #Delta=w-w0 where w is the laser frequency and w0 is  the atomic transition frequency

# Define basis states
g = basis(2, 0)  # Ground state |g>, the first argument is the dimension of the Hilbert space (2 for a two-level system), and the second argument is the index of the state.
e = basis(2, 1)  # Excited state |e>

H_0=hbar*(-2*Delta)*e * e.dag()
print(H_0)
H_I=hbar*Omega/2*sx
print(H_I)
H_total=H_0+H_I
print(H_total)

# Initial state (pure excited state)
psi0 = e # The initial state psi0 is set to the excited state e.

ee = e * e.dag() #Projection operator |e><e|, representing the population in the excited state
gg = g * g.dag() # Projection operator |g><g|, representing the population in the ground state
```

The initial **inner product** between the ground state **g** and the excited state **e** is 0 + 0j, which means they are **orthogonal states**, as expected. The fact that the inner product between the evolved states remains constant and equal to the initial inner product (0 in this case) confirms that the evolution is **isometric**. An isometric evolution preserves the inner product between states, which is exactly what you observe in your results. 

Additionally, the **trace** of the **density matrices** remains **constant** at 1 throughout the evolution, further supporting the isometric nature of the evolution.

In **quantum mechanics**, the **orthogonality** of states is a fundamental property of the **Hilbert space**, which is the mathematical framework used to describe quantum systems. In a **Hilbert space**, two states are orthogonal if their inner product is zero. Orthogonality is a key concept in quantum mechanics because it allows us to distinguish between different quantum states.

In the case of a two-level system, like the one we are simulating, the ground state and the excited state are typically chosen to be orthogonal. This orthogonality ensures that the states are distinct and can be uniquely identified. It also allows for the superposition principle, where any quantum state of the system can be expressed as a linear combination of the orthogonal basis states.

Regarding **black hole evaporation**, the situation is more complex. In the context of the information paradox and the **Page curve**, the orthogonality of states is not a strict requirement. The focus is on the **entanglement** between the black hole and the emitted radiation, and how the information is encoded in this entanglement. In the case of black hole evaporation, the Hilbert space is divided into two parts: one for the black hole and one for the emitted radiation. The orthogonality of states within each subspace is still important, but the main concern is the entanglement between the two subspaces and how it evolves over time.

The **Page curve** describes the entropy of the emitted radiation, which is related to the entanglement between the black hole and the radiation. The curve shows that the entropy initially increases, reaches a maximum, and then decreases, indicating that information is not lost but rather encoded in the entanglement.

The system undergoes **non-unitary isometric evolution**. The increase in **entropy**, decrease in **purity**, and constant **trace** suggest that the system loses information to the environment while preserving the trace, consistent with non-unitary isometric evolution. The system and its environment, when considered together, could exhibit a **Page curve**-like behavior. In this case, the total entropy of the system plus environment would follow the Page curve, while the system alone would show non-unitary isometric evolution. This would mean that information is conserved in the combined system but appears to be lost when considering only the subsystem of interest.
