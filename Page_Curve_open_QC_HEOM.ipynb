## Introduction

### Unitary and Non-Unitary Isometric Time Evolution

**Unitary** time evolution is a fundamental concept in quantum mechanics. It means that the state of a quantum system evolves according to a **unitary operator**. This operator is reversible and preserves the total probability (i.e., the sum of the probabilities of all possible outcomes remains 1). It is a way to describe how a **closed** quantum system changes over time without losing information.

**Non-unitary isometric time evolution**on the other hand, is a bit more complex. It's used when dealing with **open quantum systems**, which can interact with their environment. In this case, the evolution of the system is described by an isometric but non-unitary operator. This means that while the transformation still preserves the norm (i.e., the total probability remains 1), it is not necessarily reversible. This **non-reversibility** reflects that information can be lost to the environment in an **open system**.

A **quantum open system** refers to a quantum system that interacts with an external environment (often called a "bath" or "reservoir"). Because of this interaction, the system is not isolated, and the environment influences its dynamics. This leads to phenomena such as decoherence (loss of quantum coherence due to interaction with the environment), dissipation (loss of energy from the system to the environment) and noise (random fluctuations introduced by the environment). In open quantum systems, the evolution of the system is typically **non-unitary** because information (e.g., quantum states or energy) is lost to the environment. This is described mathematically using tools like the e.g. master equation like **Lindblad equation**. **Non-unitary isometric evolution** refers to a specific type of mathematical description used in quantum mechanics. The combined system (system + environment) evolves **unitarily**, but the evolution of the system alone is non-unitary because it is "traced over" the environment (i.e., the environment is treated as an external degree of freedom that is not observed).

In **open quantum systems**, the evolution is typically **non-unitary** because the system can exchange information and energy with its environment. This is in contrast to **closed systems**, where the evolution is **unitary** and no information is lost. When we consider the **combined system** of the quantum system and its environment, the evolution is unitary. However, when we focus only on the system and "trace over" the environment (i.e., average over all possible states of the environment), the resulting evolution for the system alone is **non-unitary**. This reflects the fact that we're not keeping track of the environment's degrees of freedom, which can carry away information from the system.

The **Page curve** is a concept from quantum information theory that describes how the **entanglement entropy** of a subsystem evolves when a **pure state** is subjected to **unitary evolution**. In simpler terms, it shows how the amount of information shared between two parts of a system changes as the system evolves. The curve is named after Don Page, who first described it in the context of **black hole evaporation**. The **Page curve** typically has a characteristic shape: it starts at zero, rises to a maximum value, and then decreases back to zero. This behavior is thought to be a consequence of the subsystem becoming increasingly **entangled** with the rest of the system, and then becoming less entangled as the system approaches a pure state again.

The **Page curve**, which describes the entropy dynamics of a **black hole**, provides a framework for understanding how information is preserved during **black hole evaporation**. According to the Page curve, the entropy of a black hole initially increases as it emits Hawking radiation, reaching a maximum value at the Page time. After the **Page time**, the entropy starts to decrease: the emitted radiation carries the information away. This decrease in entropy suggests that the evaporation process is **unitary** and that information is not lost, but rather encoded in the **entanglement** between the black hole and the radiation.

### The HEOM Method

The **Hierarchical Equations of Motion (HEOM) method** could be a more suitable approach for describing the evolution of Planck-scale entanglement in **black hole evaporation** compared to the **Lindblad Master Equation**. This is because the HEOM method can handle both non-Markovian dynamics and strong system-environment coupling, which are expected to be relevant in the context of black hole evaporation.


The HEOM method, a technique derived by Yoshitaka Tanimura and Ryogo Kubo in 1989, is a non-perturbative approach developed to study the evolution of a **density matrix** of quantum dissipative systems. While not directly derived from the **master equation**, it is also based on the quantum mathematical formalism. It is a numerical technique that employs a hierarchy of equations to describe the time evolution of the system’s reduced density matrix and a set of auxiliary density matrices. The HEOM method can handle a wider range of system-environment interactions compared to the master equation, including non-Markovian and non-perturbative cases.

We can also investigate the evolution of the two-level system’s **von Neumann entropy** and its relation to the **Page curve** by employing the **Hierarchical Equations of Motion (HEOM) method** in **QuTiP** [1].

A two-level quantum system (spin) interacts with a bosonic environment (bath). The **spin-boson model** is a widely used model for studying open quantum systems and their dynamics. We employ a **Drude-Lorentz spectral density** to characterize the properties of the **bosonic bath** and its coupling to the system [2]. The HEOM method allows for an accurate simulation of the time evolution of the system’s density matrix, considering the **non-Markovian** and non-perturbative effects arising from the system-bath interaction.

## Two-level system evolution

We define the basis states for a two-level quantum system. The **basis()** function is used to create the basis states. The first argument 2 represents the dimension of the **Hilbert space**, which is 2 for a two-level system. The second argument is the index of the state, where 0 represents the ground state |g> and 1 represents the excited state |e>.

</font>


```python
# Define basis states
g = basis(2, 0)  # Ground state |g>
e = basis(2, 1)  # Excited state |e>
```

<font face="Arial" size="4" color="black">

Then we define the parameters of the system. **E** represents the excited state energy, and we set it to 1.0, assuming hbar=1. **Omega** represents the **Rabi frequency** Ω, which is the strength of the interaction between the two states. In this case, it is set to 0.
</font>

```python
# Define parameters
E = 1.0     # Excited state energy (hbar=1)
Omega = 0  # Rabi frequency
```

<font face="Arial" size="4" color="black">
    
We define the Pauli matrices used in the Hamiltonian. **sz** is defined as the projection operator |e><e|, which represents the population in the excited state. It is obtained by multiplying the excited state **e** with its adjoint e.dag(). **sx** is the Pauli X operator, which is a 2x2 matrix representing the spin-1/2 particle in the x-direction. It is obtained using the sigmax() function.
</font>

```python
# Define Pauli matrices
sz = e * e.dag()  # Projection operator |e><e|
sx = sigmax()     # Pauli X operator
```

<font face="Arial" size="4" color="black">
    
We define the Hamiltonian of the system. The Hamiltonian **Hsys** is a linear combination of the projection operator sz and the Pauli X operator sx. It represents the system's evolution under the influence of the excited state energy E and the Rabi frequency Omega Ω.
</font>

```python
# Define Hamiltonian (without the 1/2 factor in Omega term)
Hsys = E * sz + Omega * sx
```

<font face="Arial" size="4" color="black">
    
Finally, we define the initial state of the system. psi0 is set to the excited state e, representing a pure excited state. We then convert the state vector psi0 to a density matrix rho0 using the ket2dm() function. The density matrix represents the state of the system in a statistical ensemble. We print the initial density matrix rho0 to see its representation.

This code sets up the basic elements of a two-level quantum system, including the basis states, parameters, Pauli matrices, Hamiltonian, and initial state. It provides a starting point for further analysis and simulations of the system's dynamics.
</font>

```python
# Initial state (pure excited state)
psi0 = e
rho0 = ket2dm(psi0)
print(rho0)
```

<font face="Arial" size="4" color="black">

The **Drude-Lorentz spectral density** and the corresponding **DrudeLorentzBath** and **DrudeLorentzPadeBath** classes in **QuTiP** are commonly used to model **bosonic environments**, such as phonons or photons, interacting with a quantum system. In the **bosonic bath model**, the environment is represented by a collection of **harmonic oscillators** with a continuous spectrum of frequencies. The Drude-Lorentz spectral density is a specific form of the spectral density function that characterizes the coupling between the system and the bosonic bath. The code provided below sets up a Drude-Lorentz environment using the DrudeLorentzEnvironment class and approximates it using the Padé expansion. The HEOMSolver is then used to solve the dynamics of the system coupled to this bosonic bath.

The **Padé expansion** has several advantages:

1. It can provide a more accurate approximation of the spectral density function compared to other expansions, such as the **Matsubara expansion**, especially for certain parameter regimes.
2. It can capture the behavior of the spectral density function over a wide range of frequencies using a relatively small number of terms.
3. It can lead to more efficient numerical simulations by reducing the number of auxiliary modes required to represent the bath. This is especially important when dealing with large-scale quantum systems or long-time simulations.
4. However, the Padé approximation may be more computationally expensive and can be sensitive to numerical instabilities in certain cases.


The **Matsubara approximation** is another commonly used method in the HEOM framework for approximating the spectral density function. It has the following characteristics:

1. The Matsubara approximation is based on the Matsubara frequencies, which are related to the poles of the Bose-Einstein distribution function.
2. It is often used in the context of **quantum dissipative systems** and can provide a good approximation of the spectral density function in many cases.
3. The Matsubara approximation is generally more computationally efficient compared to the Padé approximation.
4. However, it may require a larger number of terms to achieve the same level of accuracy as the Padé approximation, especially for certain parameter regimes.
5. It is compatible with diagrammatic techniques, such as Feynman diagrams, which are widely used in **quantum field theory** and many-body physics. This compatibility allows for a systematic perturbative treatment of interactions and the inclusion of higher-order effects within the HEOM framework.
</font>

<font face="Arial" size="4" color="black">
    
We define the parameters related to the approximation of the spectral density of the thermal bath. **Nk** represents the number of exponentials used in the approximation, and **Ncut** is the cut-off parameter for the bath
</font>

```python
Nk = 2  # number of exponentials in approximation of the spectral density
Ncut = 30  # cut off parameter for the bath
```

<font face="Arial" size="4" color="black">
    
We define the system-bath coupling operator **Q** using the Pauli X operator sigmax(). We also set the values for the coupling strength gamma, cut-off frequency lam, and temperature T of the thermal bath.
</font>

```python
Q = sigmax()
gamma = 0.05  # coupling strength
lam = 0.1  # cut off frequency
T = 0.4  # temperature
```

<font face="Arial" size="4" color="black">
    
We import the DrudeLorentzBath and DrudeLorentzPadeBath classes from the qutip.solver.heom module. These classes represent different expansions for the Drude-Lorentz spectral density of the thermal bath. For the **Matsubara expansion**, we create an instance of DrudeLorentzBath with the specified parameters: Q (coupling operator), lam (reorganization energy), gamma (cutoff frequency), T (temperature), and Nk (number of Matsubara terms). Similarly, for the **Padé expansion**, we create an instance of DrudeLorentzPadeBath with the same parameters.
</font>

```python
from qutip.solver.heom import DrudeLorentzBath
from qutip.solver.heom import DrudeLorentzPadeBath

# Matsubara expansion:
bath = DrudeLorentzBath(Q, lam, gamma, T, Nk)

# Padé expansion:
#bath = DrudeLorentzPadeBath(Q, lam, gamma, T, Nk)
```

<font face="Arial" size="4" color="black">
    
We import the DrudeLorentzEnvironment class from qutip.core.environment. We create an instance of DrudeLorentzEnvironment with the specified temperature T, cut-off frequency lam, and coupling strength gamma. We then use the approximate() method to obtain the Padé approximation of the environment with Nk exponentials. The **DrudeLorentzBath class** itself does not specify the expansion method. The expansion method is determined by the **approximate method** of the **BosonicBath** class.
</font>

```python
from qutip.core.environment import DrudeLorentzEnvironment

env = DrudeLorentzEnvironment(T, lam, gamma)

# Matsubara expansion:
approx=env.approximate("matsubara", Nk)

# Padé expansion:
#approx = env.approximate("pade", Nk)
```

<font face="Arial" size="4" color="black">
    
Finally, we import the **HEOMSolver class** from qutip.solver.heom. We set the maximum hierarchy depth, max_depth, to retain during the simulation and specify the solver options, including the number of time steps, nstep,s and whether to store the states during the evolution. We create an instance of HEOMSolver by passing the system Hamiltonian **Hsys**, the approximated environment approx along with the coupling operator **Q**, the maximum hierarchy depth max_depth, and the solver options. The HEOMSolver is initialized with the necessary parameters to solve the dynamics of the system coupled to the thermal bath.
</font>

```python
from qutip.solver.heom import HEOMSolver

max_depth = 5  # maximum hierarchy depth to retain
options = {"nsteps": 15_000, "store_states": True} 

solver = HEOMSolver(Hsys, (approx, Q), max_depth=max_depth, options=options)
```

<font face="Arial" size="4" color="black">
    
In this part of the code, we define the projection operators and run the solver to simulate the dynamics of the two-level system coupled to the thermal bath. We define the **projection operators** using the basis states **g** (ground state) and **e** (excited state). **P00** represents the population in the ground state, **P11** represents the population in the excited state, **P01** represents the coherence between the ground and excited states, and **P10** represents the coherence between the excited and ground states.
</font>

```python
P00 = g * g.dag()
P11 = e * e.dag()
P01 = g * e.dag()
P10 = e * g.dag()
```

<font face="Arial" size="4" color="black">
    
We create a time array **tlist** using np.linspace() to specify the time points at which we want to evaluate the system's dynamics. In this case, it ranges from 0 to 1000 with 1000 equally spaced points. We run the **HEOMSolver** using the run() method. We pass the initial density matrix rho0, the time array tlist, and a dictionary of expectation value operators e_ops. The keys of the dictionary are labels for the operators, and the values are the corresponding projection operators defined earlier. The solver will calculate the expectation values of these operators at each time point.
</font>

```python
tlist = np.linspace(0, 1000, 1000)
result = solver.run(rho0, tlist, e_ops={"00": P00, "11": P11, "01": P01, "10": P10})
```

<font face="Arial" size="4" color="black">

Finally, we plot the results using Matplotlib. We create a new figure using plt.figure() and plot the population of the excited state (P11) and the ground state (P00) over time. We use result.times to access the time points and np.real(result.e_data["11"]) and np.real(result.e_data["00"]) to access the real parts of the corresponding expectation values. We set the x-label to "Time", the y-label to "Population", add a legend using plt.legend(), and display the plot using plt.show().

This code simulates the dynamics of the two-level system coupled to the thermal bath using the HEOM solver and plots the population of the excited and ground states over time. The plot helps visualize the evolution of the system and observe phenomena such as relaxation and dephasing due to the interaction with the thermal bath.
</font>

```python
plt.figure()
plt.plot(result.times, np.real(result.e_data["11"]), color='red', label=r'P11')
plt.plot(result.times, np.real(result.e_data["00"]), color='blue', label=r'P00')
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.show()
```

This process raises questions about the fate of information that falls into the black hole. In quantum mechanics, unitarity ensures that the total quantum state of an isolated system is preserved over time. However, black hole evaporation seems to challenge this principle, leading to the **black hole information paradox**.

The **Page curve** illustrates the entropy of a black hole over time as it evaporates. Initially, entropy increases as the black hole emits radiation and interacts with its environment. After the Page time, entropy decreases, suggesting that information is being recovered. Isometries are maps between Hilbert spaces that preserve the inner product, ensuring that quantum states evolve without losing their essential properties. In the context of black hole evaporation, **isometries** describe how the black hole and its environment interact while preserving information.

Modeling black hole radiation with a **two-level system** (representing the black hole) and a **Drude-Lorentz bath** (representing the environment) creates an **open quantum system**. The ground and excited states of the two-level system are **orthogonal**, meaning their inner product is zero. As the system evolves, the **isometry** maps the states of the two-level system into the larger Hilbert space that includes the bath. This evolution appears non-unitary from the perspective of the two-level system alone, leading to the perception of information loss. However, the isometry ensures that the total evolution in the larger Hilbert space is unitary, preserving information through entanglement between the system and the bath.

The decrease in entropy after the **Page time** suggests that information, once scrambled and entangled with the environment, becomes unscrambled and encoded in the correlations between the black hole and the radiation. This aligns with the idea that information is not lost but preserved in the **entanglement**. **Isometries **play a crucial role in transferring information from the black hole to the radiation. By preserving the inner product and ensuring unitary evolution in the larger Hilbert space, isometries facilitate the encoding of information in the entanglement between the black hole and the environment.
